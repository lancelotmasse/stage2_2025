```python
#Importe les packages nécessaires

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
from scipy.stats import linregress
import matplotlib.dates as mdates
from scipy.interpolate import interp1d
import matplotlib.dates as mdates
from matplotlib.colors import TwoSlopeNorm

sns.set(style="whitegrid")
```

```python
#Importe les données dans le Notebook
#data = pd.read_csv("PATHTOYOURDATA", skiprows=1, parse_dates=["NAMEOFYOURTIMECOLUMN"])
data = pd.read_csv(, skiprows=1, parse_dates=[])

#Utiliser les dates comme index
data.set_index("NAMEOFYOURDATECOLUMN", inplace=True)

#Change your UTC if necessary
#data.index = data.index - pd.Timedelta(hours-5)

#Show the first row of your data
data.head()
```

```python
#Renommer les colonnes selon LBL
new_cols = []
for col in data.columns:
    if "LBL:" in col:
        # get text after 'LBL: ' and before closing parenthesis
        new_cols.append(col.split("LBL: ")[1].rstrip(")"))
    else:
        new_cols.append(col)

data.columns = new_cols
data.drop(columns=["#"], inplace=True)
data.head()
```

```python
#Select air data
air_t = data["tair"]

#Make plot
plt.figure(figsize=(12, 6))
plt.plot(air_t.index, air_t, label="Air Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.title("Air Temperature over Time")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#data = data.rename(columns={"COLUMNNAME": columndepth, "COLUMNNAME": columndepth})
data = data.rename(columns={})

Tsol = data.drop(columns=["tair", "hr"])
sorted_cols = sorted([c for c in Tsol.columns if isinstance(c, int)])
Tsol = Tsol[sorted_cols]

Tsol.head()
```

```python
plt.figure(figsize=(12, 5))

for depth in Tsol.columns:
    plt.plot(Tsol.index, Tsol[depth], label=f"{depth} cm")

plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.title("Soil Temperature Profiles")
plt.legend(title="Depth")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Prepare matrix
Z = Tsol.T.values              # depth x time
depths = Tsol.columns.astype(int)
times = Tsol.index

# Create meshgrid for pcolormesh
X, Y = np.meshgrid(times, depths)

fig, ax = plt.subplots(figsize=(15, 6))

# Heatmap
pcm = ax.pcolormesh(X, Y, Z, shading='auto', cmap='coolwarm')
ax.invert_yaxis()  # depth increases downward

# Colorbar
cbar = fig.colorbar(pcm, ax=ax)
cbar.set_label("Temperature (°C)")

# Format x-axis as dates
ax.xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=10))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H:%M'))
fig.autofmt_xdate()

# Labels and title
ax.set_xlabel("Time")
ax.set_ylabel("Depth (cm)")
ax.set_title("Soil Temperature Heatmap")

plt.tight_layout()
plt.show()
```
