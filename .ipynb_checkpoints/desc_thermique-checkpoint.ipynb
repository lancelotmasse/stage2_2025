Script to allow computing and visualisation of dummy data.

```python
#Importe les packages nécessaires

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
from scipy.stats import linregress
import matplotlib.dates as mdates
from scipy.interpolate import interp1d
import matplotlib.dates as mdates
from matplotlib.colors import TwoSlopeNorm

sns.set(style="whitegrid")
```

```python
#Importe les données dans le Notebook
data = pd.read_csv("TP3/dummydata.csv")

#Identifie la colonne date
data["datetime"] = pd.to_datetime(data["datetime"])

#Utiliser les dates comme index
data.set_index("datetime", inplace=True)

#Show the first row of your data
data.head()
```

```python
#produire graphique des température de l'air

#Sélectionner seulement les températures de l'air
Ta = data["te_air_-140_65"]

#Ta = Ta.resample("1D").mean()

#Indiquer les attributs de la figure
plt.figure(figsize=(12, 5))
plt.plot(Ta.index, Ta.values, label="Air Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Air Temperature over Time")
plt.legend()
plt.show()
```

```python
#Produire une matrice de température avec les données de température du sol

#Sélectionner toutes les colonnes sauf celle de l'air
Tsol = data.drop(columns=["te_air_-140_65"])

#Renommer les colonnes selon la profondeur
Tsol.columns = [col.split('_')[2] for col in Tsol.columns]

#Print le début du DataFrame
print(Tsol.head())
```

```python
# Prepare matrix
Z = Tsol.T.values              # depth x time
depths = Tsol.columns.astype(int)
times = Tsol.index

# Create meshgrid with real datetimes
X, Y = np.meshgrid(times, depths)

plt.figure(figsize=(15, 6))

plt.pcolormesh(X, Y, Z, shading='auto', cmap='coolwarm')
plt.gca().invert_yaxis()
plt.colorbar(label="Temperature (°C)")

# Auto date formatting
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=10))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()

plt.xlabel("Time")
plt.ylabel("Depth (cm)")
plt.title("Soil Temperature Heatmap")

plt.tight_layout()
plt.show()
```

```python
# Original data
depths = Tsol.columns.astype(int)
times = Tsol.index

# Convert datetimes → numeric for plotting
times_num = mdates.date2num(times.to_pydatetime())

# Create finer depth grid
depth_fine = np.linspace(depths.min(), depths.max(), 133)

# Interpolate vertically for each timestamp
Tsol_fine = []
for i in range(len(Tsol)):
    f = interp1d(depths, Tsol.iloc[i, :], kind='cubic')
    Tsol_fine.append(f(depth_fine))

Tsol_fine = np.array(Tsol_fine).T    # shape = depth_fine x time

# --- Create meshgrid for pcolormesh ---
X, Y = np.meshgrid(times_num, depth_fine)

# --- Plot ---
plt.figure(figsize=(15, 5))

pc = plt.pcolormesh(X, Y, Tsol_fine,
                    shading="auto",
                    cmap="coolwarm")

plt.gca().invert_yaxis()   # deeper = lower on plot

# Colorbar
plt.colorbar(pc, label="Temperature (°C)")

# X-axis formatting
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=10))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()

# Labels
plt.xlabel("Time")
plt.ylabel("Depth (cm)")
plt.title("Interpolated Soil Temperature Heatmap")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from scipy.interpolate import interp1d
from matplotlib.gridspec import GridSpec

# --- Prepare data ---
Ta = data["te_air_-140_65"]
depths = Tsol.columns.astype(int)
times = Tsol.index

# Interpolate soil temperature
depth_fine = np.linspace(depths.min(), depths.max(), 133)
Tsol_fine = []
for i in range(len(Tsol)):
    f = interp1d(depths, Tsol.iloc[i, :], kind='cubic')
    Tsol_fine.append(f(depth_fine))
Tsol_fine = np.array(Tsol_fine).T

times_num = mdates.date2num(times.to_pydatetime())
X, Y = np.meshgrid(times_num, depth_fine)

# --- Create figure with GridSpec ---
fig = plt.figure(figsize=(15, 8))
gs = GridSpec(2, 2, width_ratios=[20, 1], height_ratios=[1, 2], wspace=0.05, hspace=0.3)

# Top plot: Air temperature
ax0 = fig.add_subplot(gs[0, 0])
ax0.plot(Ta.index, Ta.values, color="tab:blue", label="Air Temperature")
ax0.set_ylabel("Temp (°C)")
ax0.set_title("Air Temperature")
ax0.legend()
ax0.grid(True)

# Bottom plot: Soil temperature heatmap
ax1 = fig.add_subplot(gs[1, 0], sharex=ax0)
pcm = ax1.pcolormesh(X, Y, Tsol_fine, shading="auto", cmap="coolwarm")
ax1.invert_yaxis()
ax1.set_ylabel("Depth (cm)")
ax1.set_title("Soil Temperature Heatmap")

# Colorbar on the right
cax = fig.add_subplot(gs[1, 1])  # lower rows
cbar = fig.colorbar(pcm, cax=cax)
cbar.set_label("Temperature (°C)")

# X-axis formatting
ax1.xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=10))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
fig.autofmt_xdate()

plt.show()
```
