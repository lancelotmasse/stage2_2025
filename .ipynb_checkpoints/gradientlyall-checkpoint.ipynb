```python
# This code is to compute lapse rate using the data collected on site
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
from scipy.stats import linregress
from windrose import WindroseAxes

sns.set(style="whitegrid")
```

```python
# Accéder et formater les données nécessaires
lyalldata = {
    "lyallbas": pd.read_csv(r"TP3/bas_ml_20251108.csv", index_col = "Date Time, GMT-05:00", skiprows=1),
    "lyallmid": pd.read_csv(r"TP3/mid_ml_20251108.csv", index_col = "Date Time, GMT-05:00", skiprows=1),
    "lyallhaut": pd.read_csv(r"TP3/top_ml_251108.csv", index_col = "Date Time, GMT+00:00", skiprows=1)
}



for station, df in lyalldata.items():
    df.index = pd.to_datetime(df.index)
    new_cols = []
    for col in df.columns:
        if "LBL:" in col:
            # get text after 'LBL: ' and before closing parenthesis
            new_cols.append(col.split("LBL: ")[1].rstrip(")"))
        else:
            new_cols.append(col)
    
    df.columns = new_cols
    df.drop(columns=["#"], inplace=True)


lyalldata["lyallbas"].rename(columns= {"t_1": 1, "t_2": 2, "t_3": 3, "t_0": 0, "t_air":"TAm", "hr_air": "hr"}, inplace =True)  
lyalldata["lyallbas"].index.name = "Date"
lyalldata["lyallmid"].rename(columns= {"tprof1": 1, "tprof2": 2, "tprof3": 3, "tsurf": 0, "tair":"TAm"}, inplace = True)  
lyalldata["lyallmid"].index.name = "Date"
lyalldata["lyallhaut"].rename(columns= {"tprof2": 1, "tprof3": 2, "tprof4": 3, "tprof1": 0, "tamb": "TAm", "hramb": "hr"}, inplace= True)
lyalldata["lyallhaut"].index = lyalldata["lyallhaut"].index - pd.Timedelta(hours=5)
lyalldata["lyallhaut"].index.name = "Date"

for station, df in lyalldata.items():
    print(f"-------{station}--------\n", df.head())
```

```python
# Accède aux données format CSV et les enregistre comme Dataframe Pandas dans un dictionnaire.
dataWX = {
    "MontStAnne": pd.read_csv(r"TP3/WxAlbert.csv", parse_dates=["Date"], index_col="Date"),
    "ErnestLaforce": pd.read_csv(r"TP3/WxErnest.csv", parse_dates=["Date"], index_col="Date") ,
    "MontAlbert": pd.read_csv(r"TP3/WxStAnne.csv", parse_dates=["Date"], index_col="Date"),
    "lyallbas": lyalldata["lyallbas"],
    "lyallmid": lyalldata["lyallmid"],
    "lyallhaut": lyalldata["lyallhaut"],
}

# Identifie l'élévation associé à chaque station
elevations = {"MontStAnne": 1145, "ErnestLaforce": 600, "MontAlbert": 230, "lyallbas": 500, "lyallmid": 600, "lyallhaut": 800}
```

```python
### Visualisation des données
for station, df in dataWX.items():
    plt.figure()
    plt.plot(df["TAm"])
    plt.title(station)
    plt.xlabel("Date")
    plt.ylabel("Air Temperature (°C)")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
#Nettoyage des données
for station, df in {k: dataWX[k] for k in ["lyallhaut", "lyallbas", "lyallmid"]}.items():
    df = df.loc["2025-11-08 18:00":]
    df = df.resample("h").mean()
    print(df)
    dataWX[station] = df
```

```python
def hourly_lapse_estimation(data, elevations):
    """
    Compute hourly lapse rate using reference stations,
    then estimate Lyall temperatures from MontAlbert as base temp.

    Parameters:
        data : dict
            Must contain MontStAnne, ErnestLaforce, MontAlbert with 'TAm'
        elevations : dict
            Elevations for all stations

    Returns:
        estimates : DataFrame with estimated temps for Lyall stations
        lapse : DataFrame with hourly slope & intercept
    """
    
    # --- Resample reference data ---
    ref_data = {st: df["TAm"].resample("h").mean() for st, df in data.items()}
    combined = pd.DataFrame(ref_data)

    # Output DF
    lyall_stations = ["lyallbas", "lyallmid", "lyallhaut"]
    estimates = pd.DataFrame(index=combined.index, columns=lyall_stations)
    lapse = pd.DataFrame(index=combined.index, columns=["slope", "intercept"])

    # Elevation of Mont Albert
    z_albert = elevations["MontAlbert"]

    for time, row in combined.iterrows():
        row_clean = row.dropna()
        if len(row_clean) < 2:
            continue

        # regression inputs
        X = np.array([elevations[st] for st in row_clean.index])
        y = row_clean.values

        slope, intercept, *_ = linregress(X, y)
        lapse.loc[time] = [slope, intercept]

        # Anchor temperature at Mont Albert
        T_albert = combined.loc[time, "MontAlbert"]

        if pd.isna(T_albert):
            continue

        # Estimate Lyall from Mont Albert
        for st in lyall_stations:
            z = elevations[st]
            dz = z - z_albert
            estimates.loc[time, st] = T_albert + slope * dz

    return estimates.astype(float), lapse.astype(float)


est_lyall, hourly_lapse = hourly_lapse_estimation(
    {k: dataWX[k] for k in ["MontStAnne", "ErnestLaforce", "MontAlbert"]},
    elevations
)

print(est_lyall)
```

```python
lyall_obs = {
    st: dataWX[st]["TAm"].resample("h").mean()
    for st in ["lyallbas", "lyallmid", "lyallhaut"]
}

lyall_obs_df = pd.DataFrame(lyall_obs)

def compute_accuracy(estimates, observations):
    metrics = {}
    for st in observations.columns:
        est = estimates[st].reindex(observations.index)
        obs = observations[st]

        mask = obs.notna() & est.notna()
        est = est[mask]
        obs = obs[mask]

        rmse = np.sqrt(np.mean((est - obs)**2))
        mae = np.mean(np.abs(est - obs))
        bias = np.mean(est - obs)

        metrics[st] = {"RMSE": rmse, "MAE": mae, "Bias": bias}

    return pd.DataFrame(metrics).T

accuracy = compute_accuracy(est_lyall, lyall_obs_df)
accuracy
```

```python
def plot_timeseries(estimates, observations):
    for st in observations.columns:
        
        # restrict estimated temperatures to the observation period
        est = estimates[st].loc[observations.index.min():observations.index.max()]

        plt.figure(figsize=(12,4))
        plt.plot(observations[st], label="Observed", linewidth=2)
        plt.plot(est, label="Estimated", alpha=0.7)

        plt.title(f"Observed vs Estimated Temperature — {st}")
        plt.ylabel("°C")
        plt.xlabel("Time")
        plt.grid(True)
        plt.legend()
        plt.show()

plot_timeseries(est_lyall, lyall_obs_df)

```

```python
def plot_scatter(estimates, observations):
    for st in observations.columns:
        est = estimates[st].reindex(observations.index)
        obs = observations[st]

        mask = obs.notna() & est.notna()
        est = est[mask]
        obs = obs[mask]

        plt.figure(figsize=(5,5))
        plt.scatter(obs, est, alpha=0.5)
        
        # 1:1 line
        minv, maxv = min(obs.min(), est.min()), max(obs.max(), est.max())
        plt.plot([minv, maxv], [minv, maxv], 'r--')

        plt.xlabel("Observed")
        plt.ylabel("Estimated")
        plt.title(f"Estimated vs Observed — {st}")
        plt.grid(True)
        plt.show()

plot_scatter(est_lyall, lyall_obs_df)
```
